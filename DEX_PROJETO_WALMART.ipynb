<h1><b><u>TESTES ESTATÍSTICOS DE HIPÓTESE</u></b></h1>


<div align = 'justify'><h5>A fim se verificar estatisticamente se as correlações encontradas anteriormente entre as variáveis Desemprego (<b>Unemployment</b>), IPC (<b>CPI</b>), Preço do Combustível (<b>Fuel_Price</b>) e Temperatura (<b>Temperature</b>) e as vendas semanais (<b>Weekly_Sales</b>) são significativas, serão realizados testes de hipótese para o coeficiente R de Pearson. O nível de significância utilizado será de 95%, ou seja $\alpha =0,05$ | P-Value < $\alpha$.

<i><b>Vendas semanais (Weekly_Sales) x Desemprego (Unemployment)</b></i><br><ul>
<li>HO = Desemprego não possui correlação com vendas semanais ( $\rho = 0$)</li>
<li>H1 = Desemprego possui correlação com vendas ( $\rho \neq 0$)</li></ul>

<i><b>Vendas semanais (Weekly_Sales) x IPC (CPI)</b></i><ul>
<li>HO = CPI não possui correlação com vendas semanais ( $\rho = 0$)</li>
<li>H1 = CPI possui correlação com vendas ( $\rho \neq 0$)</li></ul>

<i><b>Vendas semanais (Weekly_Sales) x Preço do combustível (Fuel_Price)</b></i><ul>
<li>HO = Preço do combustível não possui correlação com vendas semanais ( $\rho = 0$)</li>
<li>H1 = Preço do combustível possui correlação com vendas ( $\rho \neq 0$)</li></ul>

<i><b>Vendas semanais (Weekly_Sales) x Temperatura (Temperature)</b></i><ul>
<li>HO = Temperatura não possui correlação com vendas semanais ( $\rho = 0$)</li>
<li>H1 = Temperatura possui correlação com vendas ( $\rho \neq 0$)</li></ul></h5></div>

<i><b>Vendas semanais (Weekly_Sales) x Feriado (Holiday_Flag)</b></i><ul>
<li>HO = Feriado não possui correlação com vendas semanais ( $\rho = 0$)</li>
<li>H1 = Feriado possui correlação com vendas ( $\rho \neq 0$)</li></ul></h5></div>

```python
columns = ['Unemployment','CPI','Fuel_Price','Temperature','Holiday_Flag']
for i in columns:
    coef,p = stats.pearsonr(df.Weekly_Sales,df[i]) #aplicação do método para o cálculo do coeficiente de correlação de Pearson, e o valor p para testar a não correlação.
    print(f'Coeficiente de correlação de Pearson da variavel {i}: {coef:.2f}\nValor de \u03C1: {p}')
```

<div align = 'justify'><h5><i>A partir dos testes acima, pode-se notar que: 
</h5><ul>
<li>Para as cinco variáveis analisadas, aquela que possui uma correlação mais forte (ainda que negativamente), é a variável desemprego (<b>Unemployment</b>), com valor de -0.11, como pôde ser demostrado anteriormente na matriz de correlação com mapa de calor (<b>heat map</b>). O valor de p é suficientemente pequeno para atestar de que a correlação entre as variáveis é significativa do ponto de vista estatístico (P-Value < $\alpha$), de modo que seja possível rejeitar a hipótese nula HO.</li>
<li>A variável IPC (<b>CPI</b>), possui correlação negativa com as vendas semanais (<b>Weekly_Sales</b>), e o seu p-value implica que essa correlação é  significativa estatisticamente (P-Value < $\alpha$), a ponto de ser rejeitada a hipótese nula H0, ainda que essa correlação seja fraca.</li>
<li>A variável preço do combustível (<b>Fuel_Price</b>), possui fraca correlação com as vendas semanais (<b>Weekly_Sales</b>, e o seu p-value  de 0.44 implica que não há evidências o bastante para rejeitar a hipótese nula H0.</li>
<li>A variável temperatura (<b>Temperatura</b>), possui correlação negativa  com as vendas semanais (<b>Weekly_Sales</b>), e o seu p-value implica que há evidências o bastante para rejeitar a hipótese nula H0 (P-Value < $\alpha$), constatando que há correlação entre a temperatura e as vendas, ainda que ela seja fraca.</li>
<li>Foi possível constatar estatisticamente que, para a variável feriado (<b>Holiday_Flag</b>), que possui correlação positiva com as vendas semanais (<b>Weekly_Sales</b>), existe evidências para rejeitar a Hipótese nula H0 (P-Value < $\alpha$),  constatando que realmente existe uma diferença positiva nas vendas para as semanas em que há feriados, com relação nas que nõa há..</li></ul>

Apesar de haver correlações lineares fracas entre as variáveis, assumir que elas existem é uma premissa válida (rejeitando a hipótese nula), pois o valor de P (P-Value) das variáveis analisadas, exceto o preço do combustível (Fuel_Price), é menor do que o nível de significância adotado ($\alpha =0,05$).<br><br>
<b>Dessa forma, nesse primeiro momento, assumiremos para critério de seleção de loja para investimento e expansão a variável que possui correlação linear mais forte (variável desemprego, <b>Unemployment</b>) para a classificação das lojas.</b></div>

<h1><b><u>AGRUPAMENTO DAS VENDAS POR LOJA, PARA CADA ANO</u></b></h1>


<div align='justify'><h5><i>A seguir, para que seja possível escolher uma loja para a recomendação de investimento, os dados do conjunto de dados serão agrupados por loja, para cada ano (2010, 2011 e 2012), e seguindo o critério de: <br>
<ul><li><b>Weekly_Sales</b>: agrupamento por média</li>
<li><b>Temperature</b>: agrupamento por média</li>
<li><b>Fuel_Price</b>: agrupamento por média</li>
<li><b>CPI</b>: agrupamento por média</li>
<li><b>Unemployment</b>: agrupamento por média</li>
</b></ul></i></h5></div>


```python
df_media = df.groupby(by=['Year', 'Month','Store']).agg(np.mean).reset_index() 
df_media = df_media.groupby(by=['Year','Store']).agg(np.mean).reset_index() 
df_media.drop(labels=['Holiday_Flag','Month','Day'],axis=1,inplace=True)
df_media
```

<div align='justify'><h5><i>A fim de que seja possível calcular a porcentagem de crescimento de vendas nos últimos 3 anos, e filtrando os resultados por aqueles que forem positivos, serão feitos os seguintes cálculos:</b></ul></i></h5></div>


```python
growth=list() #lista para armazenamento do crescimento percentual de cada loja
for i in df_media['Store'].unique():
    a = df_media.query(f'Store == {i}').sort_values(by='Year')['Weekly_Sales'].pct_change() #variação percentual nas vendas por cada ano
    a.dropna(inplace=True)
    a = a+1
    a = ((a.iloc[0]*a.iloc[1])-1)*100
    growth.append(a)
```

<h1><b><u>CRESCIMENTO PERCENTUAL DA MÉDIA DE VENDAS POR LOJA</u></b></h1>

<div align='justify'><h5><i> O gráfico abaixo apresenta o percentual de aumento das médias de vendas semanais (<b>Weekly_Sales</b>) para os 3 anos (2010 a 2012), representado pela variável crescimento (<b>Growth</b>) nessa média de vendas, em ordem decrescente.</i></h5></div>

```python
df_final = df_media.groupby('Store').agg(np.mean).reset_index()
df_final['Total_Sales'] = df.groupby('Store').agg(np.sum).reset_index()['Weekly_Sales'] #cálculo de vendas totais de cada loja, no período informado pelo conjunto de dados
df_final.drop('Year',axis=1,inplace=True)

growth=pd.Series(growth)
df_final['Growth'] = growth # criação de nova coluna no dataframe, com o crescimento percentual nas vendas de 2010 a 2012
df_final.head()
```

```python
plt.figure(figsize=(30,10))
sns.barplot(x='Store',y='Growth',
            data=df_final,
            ci=None,
            order=df_final.sort_values('Growth',ascending = False).Store)
           
plt.title('Crescimento da média de vendas semanais por Loja, em order decrescente');
```

<div align = 'justify'><h5><i>A partir do gráfico acima, pode-se notar que: 
</h5><ul>
<li>No período que vai de fevereiro de 2010 a novembro de 2012, as lojas tiveram um comportamento bastante heterogêneo com relação ao crescimento médio de suas vendas;</li>
<li>40% das lojas (18 de 45) tiveram crescimento positivo no intervalo desses 2 anos, enquanto que 60% teve crescimento negativo (27 de 45)</li></ul></div>


<h1><b><u>CRESCIMENTO POSITIVO DA MÉDIA DE VENDAS POR LOJA</u></b></h1>

<div align='justify'><h5><i> O gráfico abaixo apresenta o percentual de aumento das médias de vendas semanais (<b>Weekly_Sales</b>) para os 3 anos (2010 a 2012), representado pela variável crescimento (<b>Growth</b>) nessa média de vendas, em ordem decrescente.</i></h5></div>
